```python
def even(num):
    even_num=[]
    x=len(num)
    for i in range(x):
        if num[i]%2==0:
            even_num.append(num[i])
        else:
            continue
    print(f""""the even numbers from the given list are ' {even_num} '""")
lst = []
n = int(input("Enter number of elements: ")) 
for i in range(0, n):
    ele = int(input(f"Enter the {i+1} element"))
    lst.append(ele)  
print(lst)
even(lst)
```

```python
def reve(st1):
    st2=st1.split()
    st2.reverse()
    st1=' '.join(st2)
    print(f"""The reversed string is ' {st1} '""")
reve(input("entere string"))
```

```python
def reve(num0):
    num=abs(num0)
    num2=str(num)
    num3=list(num2)
    num3.reverse()
    num1="".join(num3)
    if (num1==num2)==True:
        print(f"""The ' {num0} ' numer is palindrome""")
    else:
        print(f"""The ' {num0} ' numer is not palindrome""")
reve(int(input("entere number")))
```

```python

def square_list(x):
    """
    this function take number and return list of all integer
    numbers till that
    """
    lis1=[]
    i=1
    while i <=x:
        lis1.append(i**2)
        i+=1
    print(f"""The square list is {lis1}""")

square_list(int(input("Please enter the max. no in square list")))
```
