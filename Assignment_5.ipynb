```python
def even_numbers(a):
    x=[]
    for i in range(1,a+1):
        if i%2==0:
           x.append(i)
        else:
            continue
    print(f"the even numbers are {x}")
even_numbers(int(input("Enter the range of numbers: ")))
```

```python
def count_vowels(s):
    counts = {i:0 for i in 'aeiouAEIOU'}
    for char in s:
       if char in counts:
           counts[char] += 1
       else:
           continue
    print(counts)
count_vowels(input("Sentence: "))
```

```python
def is_prime(s):
    z=True
    for i in range(2,s):
        if s%i==0:
            z=False
            break
        else:
            continue
    if z==True:      
        print(f"""the number "{s}" is prime number""")
    else:
        print(f"""the number "{s}" is not prime number""")
is_prime(int(input("enter the number")))
```

```python
def fibonacci(x):   
    n1, n2 = 0, 1
    count = 0
    if x <= 0:
       print("Please enter a positive integer")
    elif x == 1:
       print("Fibonacci sequence upto",x,":")
       print(n1)
    else:
       print("Fibonacci sequence:")
       while count < x:
           print(n1)
           nth = n1 + n2
           n1 = n2
           n2 = nth
           count += 1
fibonacci(int(input("How many terms? ")))
```

```python
def is_armstrong(number):
    digits = str(number)
    length = len(digits)
    sum_of_powers = sum(int(digit) ** length for digit in digits)
    return sum_of_powers == number
def find_armstrong_numbers(end):
    armstrong_numbers = []
    for num in range(0, end + 1):
        if is_armstrong(num):
            armstrong_numbers.append(num)
    return armstrong_numbers
n=int(input(print("Enter Number")))
armstrong_numbers = find_armstrong_numbers(n)
print(f"Armstrong numbers between 0 and {n} are: {armstrong_numbers}")
```

```python
def char_frequ(s):
    counts = {i:0 for i in s}
    for char in s:
       if char in counts:
           counts[char] += 1
       else:
           continue
    print("The characters frequency counts are:",counts)
char_frequ(input("Sentence: "))
```

```python
def is_palindrome(str1):
    str2 = str1.lower().replace(" ", "")
    str3=list(str2)
    str3.reverse()
    str5="".join(str3)
    if (str2==str5)==True:
        print(f"""The ' {str1} ' string is palindrome""")
    else:
        print(f"""The ' {str1} ' string is not palindrome""")
is_palindrome(input("entere string"))
```

```python
def find_factors(n):
    x=1
    while n>0:
        x=x*(n)
        n=n-1
    print(f"""the factorial is {x}""")       
find_factors(int(input("Enter the number"))) 
```

```python
def isperfectnum(num):
    sum=0
    for i in range(1, num ):
        if num % i == 0:
           sum=sum+i
    if sum==num:
        print(f"The number {num} is perfect number")
    else:
        print(f"The number {num} is not perfect number")
isperfectnum(int(input("enter number")))
```

```python
def common_elements(lis1,lis2):
    l1=len(lis1)
    l2=len(lis2)
    lis3=[]
    for i in range(l1):
        for j in range(l2):
            if lis1[i]==lis2[j]:
                lis3.append(lis1[i])
            else:
                continue
    print(f"The common elements between the two lists are : {lis3}")

l1=[1,10,15,"ali",500]
l2=[10,"ali",20,14,15]
common_elements(l1,l2)
```

```python
def caesar_cipher(plainText, shift): 
  cipherText = ""
  for ch in plainText:    
      stayInAlphabet = ord(ch) + shift 
      if stayInAlphabet > ord('z'):
        stayInAlphabet -= 26
      finalLetter = chr(stayInAlphabet)
      cipherText += finalLetter
  print ("Your ciphertext is: ", cipherText)
caesar_cipher(input("enter string"),int(input("enter the shifted numbers")))
```
