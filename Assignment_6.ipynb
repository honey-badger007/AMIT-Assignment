```python
class Rectangle():
    def __init__(self,length,width):
        self.length=length
        self.width=width
    
    def perimiter(self):
        self.per= 2*(self.length+self.width)
        print(f"The perimeter of rectangle with (length ={self.length}, width ={self.width}) is =",self.per )       
    
    def area(self):
        self.are=self.length*self.width
        print(f"The Area of rectangle with (length ={self.length}, width ={self.width}) is =", self.are )       
   
    def display(self):
        self.area()
        self.perimiter()
        print(f"""The rectangle length is : {self.length}\nThe rectangle width is : {self.width}
The rectangle area is :{self.are}
The rectangle perimeter is :{self.per}""")

class Parallelepipede(Rectangle):
    def __init__(self,length,width,height):
        super().__init__(length,width)
        self.height=height
    def volume(self):
        self.vol= self.length*self.width*self.height
        print(f"The volume of rectangle with (length ={self.length}, width ={self.width}, height {self.height}) is =",self.vol )       
    
rec1=Parallelepipede(10,5,20)
rec1.display()
rec1.volume()
```

```python
class Person():
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def display(self):
        print(f"The Person Name is:{self.name}\nThe person Age is:{self.age}")

class Student(Person):
    def __init__(self,name,age,grade):
        super().__init__(name,age)
        self.grade=grade
        
    def display_Student(self):
        print(f"""The Student Name is: {self.name}\nThe Student Age is: {self.age}
The Student grade is = {self.grade} """)
        
Ahmed=Person("Ahmed",35)
Ahmed.display()
Ali=Student("Ali",35,"A+")
Ali.display_Student()
```
